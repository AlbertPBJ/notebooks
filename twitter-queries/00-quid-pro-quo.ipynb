```python
src['eastern_created_at'] = src['created_at'].dt.tz_localize("GMT").dt.tz_convert('US/Eastern')
src['eastern_date_time'] = pd.to_datetime(src['eastern_created_at']).dt.strftime('%Y/%m/%d %H:%M')
src['eastern_date'] = pd.to_datetime(src['eastern_created_at']).dt.strftime('%Y/%m/%d/')
src['eastern_time'] = src['eastern_created_at'].dt.time
src.eastern_date_time = pd.to_datetime(src.eastern_date_time)
src['eastern_date'] = pd.to_datetime(src.eastern_date)
```

```python
src['year'] = src['eastern_created_at'].dt.year
src['month'] = src['eastern_created_at'].dt.month
src['day'] = src['eastern_created_at'].dt.day
src['hour'] = src['eastern_created_at'].dt.hour
src['minute'] = src['eastern_created_at'].dt.minute
```

```python
src[['id', 'user_id', 'year', 'conversation_id']] =\
src[['id', 'user_id', 'year', 'conversation_id']].astype(str)
```

```python
qpq = pd.DataFrame(src)
```

```python
len(qpq)
```

```python
today = pd.datetime.today().strftime("%m/%d/%Y")
```

```python
#from when sondland texted taylor saying "no quid pro quo"
ukraine = qpq[(qpq['date'] > '2019-09-09') & (qpq['date'] < today)]
```

```python
len(ukraine)
```

```python
ukraine.tail()
```

```python
dates = ukraine.groupby(['date', 'hour']).agg('size').reset_index(name='count')
```

```python
dates.head()
```

```python
alt.Chart(dates).mark_bar().encode(
    x=alt.X('hour:N', axis=alt.Axis(values=(0,12,23)), title="Hour"),
    y=alt.Y("count:Q", title=" ", axis=alt.Axis(tickCount=3, values=(0,4000,8000), format='~s')),
    facet=alt.Facet('date:T', title=' ')
).properties(width=100, columns=6,height=100, title='Mentions of "quid pro quo" on Twitter')
```

```python
alt.Chart(dates).mark_bar().encode(
    x=alt.X('monthdate(date):N', title="Hour"),
    y=alt.Y("count:Q", title=" ", axis=alt.Axis(tickCount=3, format=',S'))
).properties(width=850, columns=6,height=300, title='Verified mentions of "quid pro quo" on Twitter')
```
