### Get our list of retweets the @VoteNoOnHolly tweet

```python
retweets_list = api.retweets(ID, count)
```

### Create a list of retweeters by screen_name

```python
rters = []

for retweet in retweets_list: 
    rters.append(retweet.user.screen_name)
```

### Get user information for each retweeter

```python
df_users = []

for r in rters:
    screen_name = r
    user = api.get_user(screen_name)
    df_users.append(user._json)
```

### Users into a dataframe

```python
df = pd.DataFrame(df_users)
```

### Concatentate the list of people the RTers follow

```python
path = 'data/raw/following/'
files = glob.glob(os.path.join(path, "*.txt"))
```

```python
file_df = (pd.read_csv(f, encoding = "ISO-8859-1", low_memory=False,sep=" ", header=None, names=["user_followed", "user"])\
           .assign(user=os.path.basename(f)) for f in files)
```

```python
following_df = pd.concat(file_df, ignore_index=True, axis=0)
```

```python
following_df.user = following_df.user.str.replace('.txt', '')
```

```python
following_df.head()
```

```python
following_df.to_csv('data/processed/following_df.csv', index=False)
```
